```python
diversity_school = pd.read_csv('diversity_school.csv')
mask_total_minority = diversity_school['category'] == "Total Minority"
diversity_prop = diversity_school[mask_total_minority]
diversity_prop["diversity_prop"] = diversity_prop['enrollment'] / diversity_prop["total_enrollment"]
diversity_prop = diversity_prop.sort_values( by="diversity_prop", ascending=False)
mask_total_enrollment = diversity_prop['total_enrollment'] > 20000
diversity_prop2 = diversity_prop[mask_total_enrollment]

ax = sns.barplot(data = diversity_prop2[:5], x = "name", y = "diversity_prop")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set(title = "Diversity in Colleges With at Least 20,0000 Total Enrollment")
```

```python
merged_copy2 = pd.merge(IPEDS_2, diversity_prop, on = "name")
merged_copy2 = pd.merge(merged_copy2, tuition_cost, on = "name")
merged_copy2
```

Data Visualization for Hypothesis testing

```python
visualization = sns.lmplot(x = 'diversity_prop', y = 'ADM_RATE_ALL', data = merged_copy2)
```

```python
visualization = sns.lmplot(x = 'diversity_prop', y = 'SAT_AVG_ALL', data = merged_copy2)
```

I used the visualizations above to find any relationships with diveristy proportion vs Admissions rates and average SATs. I saw there was a negative linear correlation, but wanted to further investigate using hypothesis testing.

For my hypothesis testing, I am going to divide the schools into two categories, one being above 23% diveristy which is the median percentage to encapsulate 50% of our data in one catgeory while the other category is below 23% diveristy. I droped NAN for admission rates so it would not interfere with our hypothesis testing.

```python
merged_copy3 = merged_copy2.dropna(subset = ['ADM_RATE_ALL'])
merged_copy3.describe()
```

```python
merged_copy3["diversity"] = merged_copy3['diversity_prop'] > 0.23
merged_copy3
```

```python

high_diversity = (merged_copy3['diversity']==True)
low_diversity = (merged_copy3['diversity']==False)

starH = merged_copy3[high_diversity]['ADM_RATE_ALL']
starL = merged_copy3[low_diversity]['ADM_RATE_ALL']
starTest = stats.ttest_ind(starH, starL)
test = starTest[1]
print(test)
```

Results

Null Hypothesis: There is not a difference in admission rates with schools with a diveristy proportion greater than 23% and schools with a diveristy proportion less than 23%.

Alt Hypothesis: There is a difference in admission rates with schools with a diversity proportion greater than 23% and schools with a diversity proportion less than 23%.

While using a two sided ttest, I found that the p value is about 0.000000014. this indicates that it our findings are statistically significant since it is much less than level 0.05. So we can rejct the null hypothsis for the alt hypothesis.
